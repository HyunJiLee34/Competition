```
import pandas as pd
import pickle
import os
```

#모델 번호 설정

```
season=int(input("시즌입력 : ")) # 시즌3
modelNum=int(input("사용모델번호 입력: ")) 

path=f'/gdrive/My Drive/Colab Notebooks/dacon/bitcoin/season{season}/modelP/modelP_{modelNum}'
samples=os.listdir(path)
sampleSize=len(samples)

samples.sort()
print("샘플사이즈는 :",sampleSize)
```

# algo 1
# 매수량 모두 1

```
df=pd.DataFrame(columns=["sample_id","buy_quantity","sell_time"])

for sampleId in samples:
    print(sampleId)
    with open(path+f"/{sampleId}", 'rb') as fr:
        user_loaded = pickle.load(fr) # [smaple_id, 매도시기, 최고가격, 가격흐름]
        sId,maxMinute,highPrice,L=user_loaded
        print(len(L))
        if maxMinute==-1:
            buyQuantity=0
            maxMinute=0
        else:
            buyQuantity=1

        df=df.append({"sample_id":sampleId[:-7],"buy_quantity":buyQuantity,"sell_time":maxMinute},ignore_index=True)

df["sample_id"]=df["sample_id"].astype(int)
df["sell_time"]=df["sell_time"].astype(int)

display(df)
```

# algo 2
# 매도시점 10이하인것은 매수x, 나머지는 풀매수(1)

```
df=pd.DataFrame(columns=["sample_id","buy_quantity","sell_time"])

for sampleId in samples:
    print(sampleId)
    with open(path+f"/{sampleId}", 'rb') as fr:
        user_loaded = pickle.load(fr) # [smaple_id, 매도시기, 최고가격, 가격흐름]
        sId,maxMinute,highPrice,L=user_loaded

        if maxMinute<=10:
            buyQuantity=0
            maxMinute=0
        else:
            buyQuantity=1

        df=df.append({"sample_id":sampleId[:-7],"buy_quantity":buyQuantity,"sell_time":maxMinute},ignore_index=True)

df["sample_id"]=df["sample_id"].astype(int)
df["sell_time"]=df["sell_time"].astype(int)

display(df)
```

# algo3 
# 평균이 1보다 작은것 and 매도가격 1보다 큰게 없는것 =>매수x
# 매도타이밍이 10보다 작은것 => 매수x (오차)

```
df=pd.DataFrame(columns=["sample_id","buy_quantity","sell_time"])
# meanL=[]

for i in range(529):
    with open(f'/gdrive/My Drive/Colab Notebooks/dacon/bitcoin//modelP/modelP_{modelNum}/{i}.pickle', 'rb') as fr:
        print(i)
        user_loaded = pickle.load(fr) # [smaple_id, 매도시기, 최고가격, 가격흐름]
        nowMean=sum(user_loaded[3])/120

        # meanL.append(nowMean)
        if (user_loaded[1]==-1 and nowMean<1) or user_loaded[1]<10:
            buy_q=0
            sell_t=0
        else:
            buy_q=1
            sell_t=user_loaded[1]

        df=df.append({"sample_id":int(i),"buy_quantity":buy_q,"sell_time":int(sell_t)},ignore_index=True)

# print(meanL)
```
